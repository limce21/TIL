```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, '--k')
plt.plot(x, z, 'b')
plt.title("THE MATPLOTLIB")
plt.xlabel('X')
plt.ylabel('SinX')
```

figure 1개, axes 1개

그림을 두개 그려보자

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(14,8))
plt.subplot(1, 2, 1) # 격자 두개가 생성된다. 그 중에 첫번째
plt.plot(x, y)
plt.subplot(1, 2, 2) # 두번째
plt.plot(x, z)
plt.show()
```

4개 만들기

```python
plt.figure(figsize=(14,8))
plt.subplot(2, 2, 1)
plt.plot(x, y)

plt.subplot(2, 2, 2)
plt.plot(x, z)

plt.subplot(2, 2, 3)
plt.plot(x, z)
plt.show()
```

```python
plt.figure(figsize=(14,8))

plt.subplot(2, 2, 2)
plt.plot(x, z)

plt.subplot(2, 2, 3)
plt.plot(x, z)
plt.show()
```

X와 Y를 Scatter, Bar, 히스토그램 형식으로 그려보자

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(14,8))
plt.subplot(2, 2, 1)
plt.scatter(x, y)

plt.subplot(2, 2, 2) # 두번째
plt.plot(x, z)

plt.subplot(2, 2, 3)
plt.bar(x, z)

plt.subplot(2, 2, 4)
plt.hist(x)

plt.show()
```

### Seaborn

```python
import seaborn as sns

sns.__version__
```

```python
tips = sns.load_dataset("tips") # seaborn의 테스트 데이터
type(tips) # 항상 무슨 타입인지 먼저 확인하자
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y="tip", hue="sex") # data: 데이터프레임, x: x축 이름, y: y축 이름, hue: 구분
plt.show()
```

```python
sns.barplot(data=tips, x='sex', y='tip')
```

```python
sns.countplot(data=tips, x='smoker') # 흡연자 팀, 아닌 팀 각각 카운트
```

```python
sns.histplot(data=tips, x='total_bill')
```

```python
tips['day'].value_counts()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill') # box plot -> 사분위값
```

사각형이 길어진다

=> 데이터가 넓게 분포되어 있다. (퍼진 정도)

25% ~ 75% 범위를 넘어가는 부분은 이상치로 친다.
